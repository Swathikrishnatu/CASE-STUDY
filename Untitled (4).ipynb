# 1. Read the dataset to the python environment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("cars_data.csv")
data.head()
```

```python
data.tail()
```

# 2. Check for the null values present in the dataset.

```python
print("The no:of NULL values present in data set")
data.isnull().sum().sum()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
df=pd.DataFrame(data)
```

```python
df["Buyer Gender"].value_counts()
```

```python
df["Buyer Gender"].value_counts().plot(kind='bar')
plt.title("Bar graph of male vs female buyers participated in the sales.")
plt.xlabel("Number of Buyers")
plt.ylabel("Buyer Gender")
plt.show()
```

# 4. Find the top 5 cars based on their sales price
